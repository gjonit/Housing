```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#reading the data
data2005 = pd.read_csv('DOF__Summary_of_Neighborhood_Sales_in_Queens_for_Class_1-__2-_and_3-Family_homes_-_2005 (1).csv')
data2006 = pd.read_csv('DOF__Summary_of_Neighborhood_Sales_in_Queens_for_Class_1-__2-_and_3-Family_homes_-_2006.csv')
data2007 = pd.read_csv('DOF__Summary_of_Neighborhood_Sales_in_Queens_for_Class_1-__2-_and_3-Family_homes_-_2007.csv')
data2008 = pd.read_csv('DOF__Summary_of_Neighborhood_Sales_in_Queens_for_Class_1-__2-_and_3-Family_homes_-_2008.csv')
data2009 = pd.read_csv('DOF__Summary_of_Neighborhood_Sales_for_Queens_for_Class_1-__2-_and_3-Family_homes_-_2009.csv')
data = [data2005,data2006,data2007,data2008,data2009]
```

```python
avg = []
low = []
high = []
sales = []
plt.figure(figsize=(20, 5))
for dat in data:
    sales.append(sum(dat['NUMBER OF SALES']))
    avg.append(dat['AVERAGE SALE PRICE'].mean())
    low.append(dat['LOWEST SALE PRICE'].mean())
    high.append(dat['HIGHEST SALE PRICE'].mean()) 
#Plotting the MEAN of Average price each year    
plt.subplot(1,4,1)   
plt.plot(avg,marker='o')
plt.title("Mean of Average Sale Price")  
#Plotting the MEAN of Lowest price each year
plt.subplot(1,4,2)
plt.plot(low,marker='o')
plt.title("Mean of Lowest Sale Price ")
#Plotting the MEAN of Highest price each year
plt.subplot(1,4,3)
plt.plot(high,marker='o')
plt.title("Mean of Highest Sale Price")
#plotting the number of sales of each year
plt.subplot(1,4,4)
plt.plot(sales,marker='o')
plt.title("Sales")
```

```python
d_one =[]
d_two = []
d_three = []
for dat in data:
    index = 0
    for i in range(0,len(dat)):
        if dat['TYPE OF HOME'][i] ==   '01 ONE FAMILY HOMES                        ':
            d_one.append(dat['NUMBER OF SALES'][i])
        elif dat['TYPE OF HOME'][i] == '02 TWO FAMILY HOMES                        ':
            d_two.append(dat['NUMBER OF SALES'][i])
        elif dat['TYPE OF HOME'][i] == '03 THREE FAMILY HOMES                      ':
            d_three.append(dat['NUMBER OF SALES'][i])
        else:
            continue
    index+=1

print("In total there are {0} One family houses sold".format(sum(d_one)))
print("In total there are {0} Two family houses sold".format(sum(d_two)))
print("In total there are {0} Three family houses sold".format(sum(d_three)))
```

```python
#Here will be shown the three different types of houses and their sales
#Will be shown the relationship betweem the Number of sales of each year with the Avergae price
year = 2005
arr = []
for i in data:
    print("YEAR: ",year)
    plt.figure(figsize=(8, 5)) 
    maxsale = max(i['NUMBER OF SALES'])
    MAXSALE_I = np.argmax(i['NUMBER OF SALES'])
    print("Max Sale:",maxsale," from ", i['TYPE OF HOME'][MAXSALE_I])
    arr.append(i['TYPE OF HOME'].value_counts())    
    print(i['TYPE OF HOME'].value_counts())
    plt.scatter(i['NUMBER OF SALES'],i['AVERAGE SALE PRICE'])
    plt.xlim(-10, 800)    
    plt.show()
    print("-----")
    year+=1
```

```python
#Average Sale price of each neighborhood for each year.
av_price =[]
nam  = []
year = 2005
inde = 1
plt.figure(figsize=(20, 5))
for dat in data:
    index = 0 
    for i in dat["TYPE OF HOME"]:
        if i == '01 ONE FAMILY HOMES                        ':
            av_price.append(dat['AVERAGE SALE PRICE'][index])
            s = dat['NEIGHBORHOOD'][index]
            s.split()
            ' '.join(s)
            nam.append(s)
        else:
            continue
        index += 1
    o = 1
    plt.subplot(1,5,inde)
    plt.xticks(fontsize=8, rotation= 90)
    plt.title("{0}".format(year))
    plt.bar(nam,av_price,0.5)
    year +=1
    inde += 1
plt.show()
```

```python

#Type of House compared with the Highests Sale Price
y = 2005
ind = 1
plt.figure(figsize=(20, 5))
for dat in data:    
    print("For the year: ", y)
    minn = np.argmin(dat['HIGHEST SALE PRICE'])
    maxx = np.argmax(dat['HIGHEST SALE PRICE'])
    print("Min Sale: ", min(dat['HIGHEST SALE PRICE']), "which is a",dat['TYPE OF HOME'][minn])
    print("Max Sale: ", max(dat['HIGHEST SALE PRICE']), "which is a",dat['TYPE OF HOME'][maxx])
    print()
    
    plt.subplot(1,5,ind)
    plt.xticks(fontsize=8, rotation= 90)
    plt.title("{0} Home Type with High Sale".format(y),fontsize=10)
    plt.scatter(dat['TYPE OF HOME'], dat['HIGHEST SALE PRICE'])
    y+=1
    ind+=1
plt.show()
```

```python
#Type of House compared with the LOWEST Sale Price

ind = 1
y = 2005
plt.figure(figsize=(20, 5))
for dat in data:    
    print("For the year: ", y)
    minn = np.argmin(dat['LOWEST SALE PRICE'])
    maxx = np.argmax(dat['LOWEST SALE PRICE'])
    print("Min Sale: ", min(dat['LOWEST SALE PRICE']), "which is a",dat['TYPE OF HOME'][minn])
    print("Max Sale: ", max(dat['LOWEST SALE PRICE']), "which is a",dat['TYPE OF HOME'][maxx])
    print()
    plt.subplot(1,5,ind)
    plt.xticks(fontsize=8, rotation= 90)
    plt.title("{0} Home Type with Lowest Sale".format(y),fontsize=10)
    plt.scatter(dat['TYPE OF HOME'], dat['LOWEST SALE PRICE'])
    y+=1
    ind+=1
plt.show()
```

```python
#Type of House compared with the AVERAGE Sale Price

ind = 1
y = 2005
plt.figure(figsize=(20, 5))
for dat in data:    
    print("For the year: ", y)
    minn = np.argmin(dat['AVERAGE SALE PRICE'])
    maxx = np.argmax(dat['AVERAGE SALE PRICE'])
    print("Min Sale: ", min(dat['AVERAGE SALE PRICE']), "which is a",dat['TYPE OF HOME'][minn])
    print("Max Sale: ", max(dat['AVERAGE SALE PRICE']), "which is a",dat['TYPE OF HOME'][maxx])
    print()
    plt.subplot(1,5,ind)
    plt.xticks(fontsize=8, rotation= 90)
    plt.title("{0} Home Type with Low Sale".format(y),fontsize=10)
    plt.scatter(dat['TYPE OF HOME'], dat['AVERAGE SALE PRICE'])
    y+=1
    ind+=1
plt.show()
```
